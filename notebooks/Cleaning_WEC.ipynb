```python
# ENERGY Lista de columnas a mantener
columns_to_keep = ["iso_code", "country", "year", "population", "per_capita_electricity", "energy_per_capita", 
                   "fossil_energy_per_capita", "renewables_energy_per_capita", "low_carbon_energy_per_capita",
                     "hydro_energy_per_capita", "solar_energy_per_capita", "wind_energy_per_capita", "other_renewables_energy_per_capita", "nuclear_energy_per_capita" 
                    ]

# #ELIMINO COLUMNAS MANTENIENDO SOLO LAS QUE QUIERO
df_energy = df[columns_to_keep]
df_energy
```

```python
df_energy[df_energy['country'] == 'Spain'].tail()
```

```python
df_energy['total_energy'] = df_energy['energy_per_capita'] * df_energy['population'] # Calcula la energía total multiplicando la energía per cápita por la población
```

```python
df_energy['total_electricity'] = df_energy['per_capita_electricity'] * df_energy['population'] # Calculo la electricidad total multiplicando la electricidad per cápita por la población
```

```python
df_energy['total_fossil_energy'] = df_energy['fossil_energy_per_capita'] * df_energy['population'] # Calculo la energia fosil total multiplicando la energia fosil per cápita por la población
```

```python
df_energy['total_renewable_energy'] = df_energy['low_carbon_energy_per_capita'] * df_energy['population'] # Calculo la energia renovable total multiplicando la low carbon (renovable + nuclear) per cápita por la población
```

```python
df_energy[df_energy['country'] == 'Spain'].tail()
```

```python
# Creo una columna con el porcentaje de energía renovable
df_energy['renewable_energy_percentage'] = (df_energy['total_renewable_energy'] / df_energy['total_energy']) * 100

# Creo una columna con el porcentaje de energía fósil
df_energy['fossil_energy_percentage'] = (df_energy['total_fossil_energy'] / df_energy['total_energy']) * 100
```

```python
df_energy['renewable_energy_percentage'] = df_energy['renewable_energy_percentage'].round(2)   # Redondeo a dos decimales

# Formateo como cadena con dos decimales
df_energy['renewable_energy_percentage'] = df_energy['renewable_energy_percentage'].apply(lambda x: '{:.2f}%'.format(x))
df_energy.head()
```

```python
df_energy['fossil_energy_percentage'] = df_energy['fossil_energy_percentage'].round(2)   # Redondeo a dos decimales

# Formateo como cadena con dos decimales
df_energy['fossil_energy_percentage'] = df_energy['fossil_energy_percentage'].apply(lambda x: '{:.2f}%'.format(x))
df_energy.head()
```
